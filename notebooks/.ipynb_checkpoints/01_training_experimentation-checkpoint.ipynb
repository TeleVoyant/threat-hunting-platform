I use `training.synthetic.generate_dataset` — the same bootstrap generator the auto-retrain

```python
from training.synthetic import generate_dataset

DURATION_HOURS = 24
HOSTS = [f"LAPTOP-{i:03d}" for i in range(1, 5)]

t0 = time.time()
labeled = generate_dataset(
    duration_hours=DURATION_HOURS,
    hosts=HOSTS,
    lateral_attacks_per_day=8,
    dns_attacks_per_day=8,
    seed=42,
)
gen_s = time.time() - t0

n_pos = sum(lbl for _, lbl in labeled)
print(f"generated {len(labeled):,} events in {gen_s:.2f}s")
print(f"  attack-labelled events : {n_pos:,} ({100*n_pos/len(labeled):.2f}%)")
print(f"  benign events          : {len(labeled)-n_pos:,}")
```

### 1a · What do the events look like?

A quick peek at the event-type distribution so I understand the raw signal mix before any
feature extraction. This is the same `NormalizedEvent` schema that flows out of the
preprocessor at runtime.

```python
rows = []
for ev, lbl in labeled:
    rows.append({
        "event_type": ev.event_type,
        "win_eid": ev.windows_event_id,
        "label": lbl,
    })
df_ev = pd.DataFrame(rows)

dist = (df_ev.groupby("event_type")
              .agg(count=("label", "size"), attack=("label", "sum"))
              .sort_values("count", ascending=False))
dist["attack_rate"] = (dist["attack"] / dist["count"]).round(3)
dist
```

**Observation.** Attack signal is spread thinly across `authentication`, `process`,
`network`, `process_access` and `dns_query` events — exactly the cross-domain mix the
detectors must disentangle. No single event type is purely malicious, which is what makes the
feature-domain restriction (Section 3) matter.

## 3 · KEY EXPERIMENT — Feature-domain restriction (improvement #1)

**Hypothesis.** Training the DNS detector on *all 156 features* lets XGBoost latch onto a
**cross-domain artefact** of the synthetic generator instead of genuine DNS signal. The
training-workflow notes record that this exact failure scored ROC-AUC **0.057** on real CIC
data while looking perfect on synthetic.

**Method.** Train two DNS models on the *same windows / same labels*:
- **(A) all 156 features**
- **(B) restricted to `dns + temporal`** (~55 features) — the production config in
  `config/detectors.yml`

Then compare eval AUC **and**, more importantly, the **feature-importance distribution** —
does the model rely on DNS features, or on something like `behavioral__*` / `process__*`?

```python
from training.trainer import train_booster

def train_on_subset(X, y, names, allowed_namespaces):
    if allowed_namespaces is None:
        keep = list(range(len(names)))
    else:
        allowed = set(allowed_namespaces)
        keep = [i for i, n in enumerate(names) if n.split("__")[0] in allowed]
    Xr = X[:, keep]
    namesr = [names[i] for i in keep]
    booster, metrics = train_booster(Xr, y, namesr, num_boost_round=120, random_state=42)
    return booster, metrics, namesr

# (A) all features
fg = load_feature_groups()
print("config feature_groups:", fg)

bst_all, m_all, names_all = train_on_subset(X_dns_full, y_dns, names_full, None)
# (B) dns + temporal (production restriction)
bst_dns, m_dns, names_dns = train_on_subset(X_dns_full, y_dns, names_full, fg["dns_exfiltration"])

print(f"\n(A) all 156 feats  -> eval_auc={m_all['eval_auc']:.4f}  feats={m_all['feature_count']}")
print(f"(B) dns+temporal   -> eval_auc={m_dns['eval_auc']:.4f}  feats={m_dns['feature_count']}")
```

### 3a · Where does each model "look"? (gain importance)

The AUC numbers alone don't tell the story — on *synthetic* data the all-features model can
look fine. The danger is *what* it relies on. I pull the top features by **gain** for each
model and flag any feature that is **not** in the DNS domain.

```python
def top_importance(booster, k=12):
    score = booster.get_score(importance_type="gain")
    s = pd.Series(score).sort_values(ascending=False).head(k)
    out = s.rename("gain").to_frame()
    out["namespace"] = [i.split("__")[0] for i in out.index]
    out["out_of_domain"] = out["namespace"].apply(lambda x: x not in {"dns", "temporal"})
    return out

print("=== (A) all-features DNS model — top gain ===")
imp_all = top_importance(bst_all)
display(imp_all)

leak = imp_all["out_of_domain"].sum()
print(f"\n>>> {leak}/{len(imp_all)} of the top features are OUT of the DNS domain "
      f"(auth/behavioral/process/network) — this is the artefact-leakage the restriction prevents.")
```

```python
print("=== (B) dns+temporal restricted DNS model — top gain ===")
imp_dns = top_importance(bst_dns)
display(imp_dns)
print(f"\n>>> out-of-domain features in top list: {imp_dns['out_of_domain'].sum()} "
      f"(should be 0 — only dns__/temporal__ remain selectable)")
```

### 3b · Visualising the leakage

```python
# Import matplotlib now — AFTER all heavy data-gen/training (see Section 0 note).
# The Jupyter kernel auto-selects its inline backend, so figures render in-notebook.
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(13, 4.5))
for ax, imp, title in [(axes[0], imp_all, "(A) all 156 features"),
                       (axes[1], imp_dns, "(B) dns + temporal")]:
    colors = ["#d9534f" if o else "#5cb85c" for o in imp["out_of_domain"]]
    ax.barh(range(len(imp)), imp["gain"].values[::-1], color=colors[::-1])
    ax.set_yticks(range(len(imp)))
    ax.set_yticklabels(imp.index[::-1], fontsize=8)
    ax.set_title(f"DNS detector — {title}")
    ax.set_xlabel("gain")
fig.suptitle("Red = out-of-DNS-domain feature (artefact leakage)  ·  Green = genuine DNS/temporal signal")
plt.tight_layout(); plt.show()
```

**Conclusion (Section 3).** The all-features model spends a large share of its gain on
`behavioral__*` / `process__*` columns it has *no business* using for DNS exfiltration — a
shortcut that evaporates on real traffic. The restricted model is forced onto genuine DNS
signals (entropy, tunnelling record ratios, beaconing CV). This is why
`config/detectors.yml` pins `dns_exfiltration → [dns, temporal]`, and why the manifest schema
is **per-detector**. I keep model (B) as the DNS candidate going forward.

## 4 · Hyperparameter tuning (Step 2 — `GridSearchCV` + `StratifiedKFold`)

Now I tune the DNS model on its **restricted** feature matrix (so the tuned params reflect the
schema the deployed model actually trains on — matching `training/tuning.py`). I optimise
`roc_auc`, the NFR-02 headline metric, because it's threshold-independent and handles the heavy
class imbalance better than accuracy.

I use a small grid here so the notebook stays interactive; the production CLI
(`python -m training.tuning`) uses the same `tune()` function.

```python
from training.tuning import tune

# Restricted DNS matrix (reuse the keep-mask from Section 3)
allowed = set(fg["dns_exfiltration"])
keep = [i for i, n in enumerate(names_full) if n.split("__")[0] in allowed]
X_dns = X_dns_full[:, keep]
names_dns_only = [names_full[i] for i in keep]
print("tuning matrix:", X_dns.shape, "positives:", int(y_dns.sum()))

# A compact grid for the live notebook (production default is 36 candidates)
grid = {
    "max_depth":        [4, 6],
    "learning_rate":    [0.05, 0.1],
    "min_child_weight": [1, 5],
    "subsample":        [0.8],
}

# n_jobs=1: single-threaded CV. With OMP_NUM_THREADS=1 the joblib-parallel path
# (n_jobs=-1) can SIGSEGV under nested OpenMP — single-threaded is the safe sandbox path.
result = tune(X_dns, y_dns, names_dns_only, param_grid=grid, n_splits=4,
              scoring="roc_auc", n_jobs=1)
print(f"\nbest CV roc_auc : {result['best_score']:.4f}")
print(f"best params     : {result['best_params']}")
print(f"candidates       : {result['n_candidates']}  ·  elapsed {result['elapsed_seconds']}s")
```

### 4a · Full sweep table

```python
cv = pd.DataFrame(result["cv_results"])
cv_params = pd.json_normalize(cv["params"])
cv_tab = pd.concat([cv[["rank", "mean_score", "std_score", "fit_time"]], cv_params], axis=1)
cv_tab = cv_tab.sort_values("rank").reset_index(drop=True)
cv_tab.round(4)
```

I persist the winning params to `data/tuning/` in exactly the JSON shape the production
`tuning.py` writes, so I could feed it straight back via
`train_models.py --params-json`.

```python
from pathlib import Path
result["model_name"] = "dns_exfiltration"
result["window_minutes"] = WINDOW_MINUTES
out_path = Path("notebooks/artifacts/dns_tuning_result.json")
out_path.write_text(json.dumps(result, indent=2, default=str))
print("wrote", out_path)
print("  -> production usage: python -m training.train_models --params-json", out_path)
```

## 5 · Train the final DNS model with tuned params and evaluate (NFR-02)

I retrain with the winning hyperparameters and evaluate using the **same metric functions** as
`training/evaluate_models.py`. The NFR-02 targets are the acceptance criteria my supervisor
cares about.

```python
from training import evaluate_models as ev

# Retrain DNS with tuned params on the restricted matrix
best = dict(result["best_params"])
xgb_params = {
    "max_depth": best["max_depth"],
    "eta": best["learning_rate"],
    "min_child_weight": best["min_child_weight"],
    "subsample": best["subsample"],
}
bst_final, m_final = train_booster(X_dns, y_dns, names_dns_only,
                                   num_boost_round=200, params=xgb_params, random_state=42)
print("final DNS model eval_auc:", round(m_final["eval_auc"], 4))

# Score the full matrix for a holistic evaluation curve set
dmat = xgb.DMatrix(X_dns, feature_names=names_dns_only)
y_score = bst_final.predict(dmat)
y_true = y_dns.astype(int)
```

```python
# NFR-02 metric block — mirrors evaluate_models.py
sweep = ev.threshold_sweep(y_true, y_score)
roc = ev.roc_auc(y_true, y_score)
prauc = ev.pr_auc(y_true, y_score)

# pick operating point at threshold 0.5
op = min(sweep, key=lambda r: abs(r["threshold"] - 0.5))
metrics_at_05 = {
    "roc_auc": round(roc, 4),
    "pr_auc": round(prauc, 4),
    "precision@0.5": round(op["precision"], 4),
    "recall@0.5": round(op["recall"], 4),
    "f1@0.5": round(op["f1"], 4),
    "fpr@0.5": round(op["fpr"], 4),
}
print(json.dumps(metrics_at_05, indent=2))

grade = ev.grade({"precision": op["precision"], "recall": op["recall"],
                  "roc_auc": roc, "fpr": op["fpr"]})
print("\nNFR-02 grade:", json.dumps(grade, indent=2, default=str))
```

### 5a · ROC, PR and threshold-sweep curves

```python
roc_pts = pd.DataFrame(ev.roc_curve_points(y_true, y_score))
pr_pts  = pd.DataFrame(ev.pr_curve_points(y_true, y_score))
sw      = pd.DataFrame(sweep)

fig, axes = plt.subplots(1, 3, figsize=(15, 4.2))
axes[0].plot(roc_pts["fpr"], roc_pts["tpr"], marker=".", ms=3)
axes[0].plot([0, 1], [0, 1], "--", color="grey", lw=.8)
axes[0].set_title(f"ROC (AUC={roc:.3f})"); axes[0].set_xlabel("FPR"); axes[0].set_ylabel("TPR")

axes[1].plot(pr_pts["recall"], pr_pts["precision"], marker=".", ms=3, color="#d9534f")
axes[1].set_title(f"Precision-Recall (AUC={prauc:.3f})"); axes[1].set_xlabel("recall"); axes[1].set_ylabel("precision")

axes[2].plot(sw["threshold"], sw["precision"], label="precision")
axes[2].plot(sw["threshold"], sw["recall"], label="recall")
axes[2].plot(sw["threshold"], sw["f1"], label="f1")
axes[2].axvline(0.5, ls="--", color="grey", lw=.8)
axes[2].set_title("Threshold sweep"); axes[2].set_xlabel("threshold"); axes[2].legend()
plt.tight_layout(); plt.show()
```

**Note on calibration (honest caveat for the write-up).** As recorded in
`training-workflow.md` Step 9, a *synthetic-trained* model ranks well (high AUC) but stays
**mis-calibrated** on real data — recall at a fixed 0.5 threshold compresses. Closing that gap
needs real-data training (an apt29-trained lateral model reached recall 1.0) and/or
per-detector threshold recalibration. I flag this so the AUC here is not over-read.

## 7 · Training progress — XGBoost learning curves

My supervisor asked to *see the training progress*, not just the final scores. `train_booster`
captures a per-round `evals_result` internally (train + eval `logloss` and `auc` at every
boosting iteration) but only returns the final values. Here I run the **identical** split /
params / `scale_pos_weight` logic with that history retained, so I can plot the learning curves
and the early-stopping point.

What to look for:
- **logloss** should fall and flatten; the **train↔eval gap** is the overfitting margin.
- **AUC** should rise and plateau.
- the dashed line is XGBoost's chosen **best_iteration** (early stopping `=20` rounds of no
  eval improvement) — the round actually used for the deployed model.

```python
from training.trainer import DEFAULT_PARAMS

def train_with_history(X, y, feature_names, *, params=None, num_boost_round=200,
                       early_stopping_rounds=20, train_split=0.8, random_state=42):
    """Mirror of trainer.train_booster, but returns the per-round evals_result too."""
    rng = np.random.default_rng(random_state)
    n = X.shape[0]
    perm = rng.permutation(n)
    cut = int(n * train_split)
    tr, ev = perm[:cut], perm[cut:]
    dtrain = xgb.DMatrix(X[tr], label=y[tr], feature_names=feature_names)
    deval  = xgb.DMatrix(X[ev], label=y[ev], feature_names=feature_names)
    pos = int(y.sum()); neg = int(len(y) - pos)
    p = dict(DEFAULT_PARAMS)
    if params:
        p.update(params)
    p["scale_pos_weight"] = (neg / pos) if pos > 0 else 1.0
    hist = {}
    bst = xgb.train(p, dtrain, num_boost_round=num_boost_round,
                    evals=[(dtrain, "train"), (deval, "eval")],
                    early_stopping_rounds=early_stopping_rounds,
                    evals_result=hist, verbose_eval=False)
    return bst, hist, int(bst.best_iteration)

# DNS with the tuned params from Section 4; lateral with regularized defaults
bst_dns_h, hist_dns, best_dns = train_with_history(X_dns, y_dns, names_dns_only, params=xgb_params)
bst_lm_h,  hist_lm,  best_lm  = train_with_history(X_lm,  y_lm,  names_lm)

print(f"DNS     best_iteration = {best_dns:3d}  | final eval AUC = {hist_dns['eval']['auc'][-1]:.4f}")
print(f"lateral best_iteration = {best_lm:3d}  | final eval AUC = {hist_lm['eval']['auc'][-1]:.4f}")
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 3, figsize=(16, 4.3))

# Panel 1 — DNS logloss (train vs eval)
r = range(len(hist_dns['train']['logloss']))
ax[0].plot(r, hist_dns['train']['logloss'], label="train")
ax[0].plot(r, hist_dns['eval']['logloss'],  label="eval")
ax[0].axvline(best_dns, ls="--", color="grey", lw=.9, label=f"best_iter={best_dns}")
ax[0].set_title("DNS detector — logloss"); ax[0].set_xlabel("boosting round")
ax[0].set_ylabel("logloss"); ax[0].legend()

# Panel 2 — DNS AUC (train vs eval)
ax[1].plot(r, hist_dns['train']['auc'], label="train")
ax[1].plot(r, hist_dns['eval']['auc'],  label="eval")
ax[1].axvline(best_dns, ls="--", color="grey", lw=.9)
ax[1].set_title("DNS detector — AUC"); ax[1].set_xlabel("boosting round")
ax[1].set_ylabel("AUC"); ax[1].legend()

# Panel 3 — eval AUC progress, both detectors overlaid
ax[2].plot(hist_dns['eval']['auc'], label=f"dns_exfiltration (best {best_dns})")
ax[2].plot(hist_lm['eval']['auc'],  label=f"lateral_movement (best {best_lm})")
ax[2].set_title("Eval AUC progress — both detectors")
ax[2].set_xlabel("boosting round"); ax[2].set_ylabel("eval AUC"); ax[2].legend()

plt.tight_layout(); plt.show()
```

**Reading the curves.** The lateral-movement model climbs to a high eval AUC within a few
rounds and early-stops — the auth/PtH/LSASS signals are strongly separable. The DNS model's eval
AUC plateaus almost immediately at a **lower, honest level (~0.79)** and early-stops too: with the
feature-domain restriction in place it can only use genuine DNS signal, so it can't cheat its way
to a perfect-but-fake curve (exactly the degenerate-AUC trap improvement #3 guards against).
Notice the DNS train logloss keeps inching down after the eval metric has flattened — that
growing train↔eval gap is precisely what early stopping (`best_iteration`) and the regularized
defaults (`colsample 0.6`, `min_child_weight 3`, `reg_lambda 2`) exist to cut off before it
becomes overfitting.
